# Simple Linear Regression - Predicting COVID-19 Case

<b>Import Libraries</b>

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline

sns.set(rc={'figure.figsize':(20.7,8.27)})
sns.set_style("dark")
sns.color_palette("dark")
plt.style.use("fivethirtyeight")
```

<b>Load Dataset</b>

```python
dataset = pd.read_csv("covid-19_case.csv")
dataset['day_of_year'] = pd.to_datetime(dataset['ObservationDate'], errors='coerce').dt.dayofyear
dataset.head(1)
```

```python
dataset.tail(1)
```

<b>Splitting the data into Train and Test</b>

Due to the small size (approx only 60 cases) of the data, the usual random split was not seen as a robust approach.

```python
X = np.array(dataset[['day_of_year']]).reshape(-1, 1) #independent variable
y_confirmed = np.array(dataset['Confirmed']) # dependent variable

#Splitting the data into Train and Test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain_confirmed, ytest_confirmed = train_test_split(X,y_confirmed,test_size=1/3, random_state=42)

xtrain = np.array(dataset[dataset['day_of_year'] < 80]['day_of_year']).reshape(-1,1)
ytrain_confirmed = dataset[dataset['day_of_year'] < 80]['Confirmed']

xtest = np.array(dataset[dataset['day_of_year'] > 80]['day_of_year']).reshape(-1,1)
ytest_confirmed = dataset[dataset['day_of_year'] > 80]['Confirmed']
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)

xtrain_poly = poly.fit_transform(xtrain)
xtest_poly = poly.transform(xtest)
```

<b>Fitting Model to Train Set</b>

```python
#Fitting simple linear regression to the Training Set
from sklearn.linear_model import LinearRegression # import library

regressor_confirmed = LinearRegression(n_jobs=2) # initiate regression object
regressor_confirmed.fit(xtrain, ytrain_confirmed) # fit the line

poly_regressor_confirmed = LinearRegression(n_jobs=2) # initiate regression object
poly_regressor_confirmed.fit(xtrain_poly, ytrain_confirmed) # fit the line
poly_regressor_confirmed.fit(X_poly, y_confirmed) # fit the line
```

<b>Predict the Test Set</b>

```python
#Predicting the prices
pred_confirmed_train = regressor_confirmed.predict(xtrain)
pred_confirmed_test = regressor_confirmed.predict(xtest)
poly_pred_confirmed_train = poly_regressor_confirmed.predict(xtrain_poly)
poly_pred_confirmed_test = poly_regressor_confirmed.predict(xtest_poly)
```

<b>Visualize the test result</b>

```python
#Visualizing the Test Results 
#plt.scatter(xtest, ytest_confirmed, color= 'red')
plt.plot(ytest_confirmed.reset_index(drop=True), color = 'red')
plt.plot(regressor_confirmed.predict(xtest), color = 'blue')

#plt.plot(ytrain_confirmed.reset_index(drop=True), color = 'red')
#plt.plot(pred_confirmed_train, color = 'blue')
plt.title("COVID-19 Confirmed Case by Linear Regression")
plt.xlabel("index")
plt.ylabel("num case")
plt.legend(['ground truth','prediction'])
plt.show()
```

```python
#Visualizing the Test Results 
plt.plot(ytest_confirmed.reset_index(drop=True), color = 'red')
plt.plot(poly_pred_confirmed_test, color = 'blue')

# plt.plot(ytrain_confirmed, color = 'red')
# plt.plot(poly_pred_confirmed_train, color = 'blue')
plt.title("COVID-19 Confirmed Case by Linear Regression")
plt.xlabel("index")
plt.ylabel("num case")
plt.legend(['ground truth','prediction'])
plt.show()
```

<b>Eval with Polynomial Feature</b>

```python
# The coefficients
print('Coefficients: \n', poly_regressor_confirmed.coef_)

# The Intercept
print('Intercept: \n', poly_regressor_confirmed.intercept_)

# Explained variance score: 1 is perfect prediction
from sklearn.metrics import r2_score
print('Variance score: %.2f' % r2_score(ytest_confirmed, poly_pred_confirmed_test))

# The mean absolute value
from sklearn.metrics import mean_absolute_error
print("MAE: %.2f" % mean_absolute_error(ytest_confirmed, poly_pred_confirmed_test))

# The mean squared error
from sklearn.metrics import mean_squared_error
print("MSE: %.2f" % mean_squared_error(ytest_confirmed, poly_pred_confirmed_test))

# RMSE
print("RMSE: %.2f" % np.sqrt(mean_squared_error(ytest_confirmed,poly_pred_confirmed_test)))
```

<b>Discussion</b>

From the given plot, the model seems underfitting. Belum bisa menyesuaikan dengan dataset yang ada. biasnya masih tinggi. maka kita butuh build model dengan kompleksitas yang lebih.

# Ridge

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge()
ridge_model.fit(xtrain, ytrain_confirmed)

poly_ridge = Ridge(alpha=4) # initiate regression object
poly_ridge.fit(xtrain_poly, ytrain_confirmed) # fit the line
```

```python
#Predicting the prices
pred_r_confirmed = ridge_model.predict(xtest)
pred_r_poly = poly_ridge.predict(xtest_poly)
```

```python
#Visualizing the Test Results 
plt.plot(ytest_confirmed.reset_index(drop=True), color = 'red')
plt.plot(pred_r_poly, color = 'blue')

plt.title("COVID-19 Confirmed Case by Linear Regression - with Polynomial Feature")
plt.xlabel("index")
plt.ylabel("num case")
plt.legend(['ground truth','prediction'])
plt.show()
```

```python
# The coefficients
print('Coefficients: \n', poly_ridge.coef_)

# The Intercept
print('Intercept: \n', poly_ridge.intercept_)

# Explained variance score: 1 is perfect prediction
from sklearn.metrics import r2_score
print('Variance score: %.2f' % r2_score(ytest_confirmed, pred_r_poly))

# The mean absolute value
from sklearn.metrics import mean_absolute_error
print("MAE: %.2f" % mean_absolute_error(ytest_confirmed, pred_r_poly))

# The mean squared error
from sklearn.metrics import mean_squared_error
print("MSE: %.2f" % mean_squared_error(ytest_confirmed, pred_r_poly))
```
